```python
import nltk
```

```python
nltk.download("punkt")
```

```python
data["Text"]
```

```python
data["num_characters"] = data["Text"].apply(len)
```

```python
data
```

```python
data["word_character"] = data["Text"].apply(lambda x:len(nltk.word_tokenize(x)))
```

```python
data
```

```python
data["Sen_count"] = data["Text"].apply(lambda x: len(nltk.sent_tokenize(x)))
```
